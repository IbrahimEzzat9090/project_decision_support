```python
import pandas as pd
import numpy as np
fname = '/content/data2.csv'
df = pd.read_csv(fname)


# print(df)
for x in range(len(df['Age'])):
  if df["Age"][x] > 81:
    df["Age"][x] = np.nan
df
# df.to_csv("\content\modified.csv",index=True)

```

```python
import math
#method 1 if empty values
# 1. delete row
# 2. create random
# disadv data outliers
# (3.) add avg

sum=0
num=0
for i in range(len(df["Age"])):
  if math.isnan(df["Age"][i]) == False:
    num += 1
    sum += int(df["Age"][i])
print(num,"faults")
print(len(df["Age"]),"data size")
avg = int(sum/num)

if len(df["Age"])!=num:
  print(f"replacing faults with {avg}")

  # for i in range(len(df["Age"])):
  #   if math.isnan(df["Age"][i]) == True:
  #     # print("fixing: ",df.loc[[i]])
  #     df.iloc[i, "Age"] = avg

  df['Age']=df['Age'].apply(lambda x:avg if(math.isnan(x)) else x)
  print("\n\n\done\n\n\n")
else:
  print("no faults")
```

```python
min=df['Room Number'][0]
max=df['Room Number'][0]
num=0
for i in range(len(df['Room Number'])):
  if math.isnan(df["Room Number"][i]) == False:
    num+=1
    if max<df['Room Number'][i]:
      max=df['Room Number'][i]
    if min>df['Room Number'][i]:
      min=df['Room Number'][i]

if len(df["Room Number"])!=num:
  print(f"Nan found on [Room Number] repalcing with {int((max+min)/2)}")
  df['Room Number']=df['Room Number'].apply(lambda x:float(int((max+min)/2)) if(math.isnan(x)) else x)

else:
  print("no data is cleaned")
```
